#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
import math

def std_dev(data):
    n = len(data)
    mean = sum(data)/n
    sd = 0
    for num in data:
        sd += ((float(num) - mean)**2)/n
    sd = math.sqrt(sd)
    return sd
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

````python
final_population_sd = {i: round(std_dev(final[i]), 2) for i in final.columns[1:]}
final_population_sd
````

Do your manually calculated standard deviations match up with the output from pandas `describe`? What about numpy's `std` method?

````python
# generate standard deviation using numpy
np.std(final)
````

The answer for standard deviation using Numpy's `std` method is similar with the answer using manually calculated standard deviation, however both's answers are different from the pandas's describe method.

Numpy's std method and manually calculated standard deviation perform the calculation based on the population standard deviation. The population standard deviation is using the size of the population to find the standard deviation.

Standard deviation from pandas's `describe` method perform the calculation based on the sample standard deviation, whereby it uses the sample as the size of the whole population and it has greater variability. Thus, the value of sample standard deviation is greather than the population standard deviation.

For the current project, we will use the sample standard deviation because we do not want to generalize the finding to the population.

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. Here's some starter code:

````python
import matplotlib.mlab as mlab
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels, list_of_ylabels):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(15,10)) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column],color='darkorange', bins=20) # feel free to add more settings
        ax[i].set_title(list_of_titles[i],fontsize=16) # Set titles for each subplot
        ax[i].set_xlabel(list_of_xlabels[i],fontsize=13) # Set xlabel for each subplot
        ax[i].set_ylabel(list_of_ylabels[i],fontsize=13)
        ax[i].axvline(dataframe[column].mean(), label='Mean', color='g', linestyle='dashed', linewidth=2)
        ax[i].axvline(np.percentile(dataframe[column],25), label='25%', color='b', linestyle='dashed', linewidth=2)
        ax[i].axvline(np.percentile(dataframe[column],50), label='50%', color='r', linestyle='dashed', linewidth=2)
        ax[i].axvline(np.percentile(dataframe[column],75), label='75%', color='b', linestyle='dashed', linewidth=2)
        ax[i].legend()
````

#### Distributions in the data

In this dataset, each data represents a sample from a population.                        
For example, for ACT math test:
- Population: the test results of all the students who take this test, nation-wide.
- Population mean: is the national average of ACT math test (total scores/total no. of test takers) 
- Sample: the state means of ACT math test. We have 51 samples (51 states)

***According to CLT, we generally assuming that data we sample from a population will be normally distributed. Do we observe this trend?***

The CLT tells us that when the sample size is increased, the distributions of the sample means will tend toward a normal distribution, even though the distribution of the population is not normally distributed. But, the data does not have the correct information to apply to CLT theorem.

We have to consider the number of the participants of each states instead of the participation rate(%) because it could affect the state's means score(sample). The participation rate varies from 0% to 100% across the states. The mean score for each state with a high participation rate should not be considered as similar to the state with a low participation rate, even if we use the same sample size across each state to get the sample mean.

Thus, in order to apply CLT theorem, it is strongly recommended to use the sample from the population(nation-wide). We should avoid using the state's mean results even though the number of participants of each state is equal because it could bias state's result.

One cannot see the normal distribution trend just by looking at the data. We should generate the histogram to observe the trend.
